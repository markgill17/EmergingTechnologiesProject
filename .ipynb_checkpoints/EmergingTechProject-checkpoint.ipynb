```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Input
from keras.utils import np_utils

from keras.regularizers import l2
```

```python
from keras.utils import np_utils
from tensorflow import keras
import keras
from keras.datasets import mnist

#creating tuples
(image_train, label_train), (image_test, label_test) = mnist.load_data()
print("Original X shape", image_train.shape)
print("Original Y shape", label_train.shape)

# Reshape data.
image_train = image_train.reshape(60000, 784)
image_test  = image_test.reshape(10000, 784)

# cast data to float type. Then we normalize it to give it a common scale
image_train = image_train.astype('float32') 
image_test  = image_test.astype('float32')
image_train /= 255 # Original data is uint8 (0-255). Scale it to range [0,1].
image_test  /= 255

print("Training X matrix shape", image_train.shape)
print("Testing X matrix shape", image_test.shape)
print (label_train[0])

# Represent the targets as one-hot vectors: e.g. 2 -> [0, 0, 1, 0, 0, 0, 0, 0, 0].
nb_classes = 10
label_train = np_utils.to_categorical(label_train, nb_classes)
label_test  = np_utils.to_categorical(label_test, nb_classes)
print("Training Y matrix shape", label_train.shape)
print("Testing Y matrix shape", label_test.shape)
```

```python
# Plot examples of the data.
plt.figure(1, figsize=(14,3))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(image_train[i].reshape(28,28), cmap='gray', interpolation='nearest')
    plt.xticks([])
    plt.yticks([])
    
```

```python
# Simple fully-connected neural network with 2 hidden layers.
model = Sequential()
model.add(Dense(256, activation='relu', input_shape=(784,), W_regularizer=l2(0.001))) # Use input_shape=(28,28) for unflattened data.
model.add(Dense(256, activation='relu'))
#model.add(Dropout(0.2)) # Including dropout layer helps avoid overfitting.
model.add(Dense(10, activation='softmax')) # Use softmax layer for multi-class problems.
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(image_train, label_train, batch_size=128, nb_epoch=10, verbose=1)
```

```python
score = model.evaluate(image_test, label_test, verbose=0)
print('Test cross-entropy loss: %0.5f' % score[0])
print('Test accuracy: %0.2f' % score[1])
```

```python
print (score)
```

```python
predicted_classes = model.predict_classes(image_test)
correct_indices   = np.nonzero(predicted_classes == label_test.argmax(axis=-1))[0]
incorrect_indices = np.nonzero(predicted_classes != label_test.argmax(axis=-1))[0]

plt.figure(1, figsize=(7,7))
for i, correct in enumerate(correct_indices[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(image_test[correct].reshape(28,28), cmap='gray', interpolation='none')
    plt.title("Predicted {}, Class {}".format(predicted_classes[correct], label_test[correct].argmax(axis=-1)))
    plt.xticks([])
    plt.yticks([])
    
plt.figure(2, figsize=(7,7))
for i, incorrect in enumerate(incorrect_indices[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(image_test[incorrect].reshape(28,28), cmap='gray', interpolation='none')
    plt.title("Predicted {}, Class {}".format(predicted_classes[incorrect], label_test[incorrect].argmax(axis=-1)))
    plt.xticks([])
    plt.yticks([])
```
